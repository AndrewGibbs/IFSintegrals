```julia
using IFSintegrals, Plots
```

Define a couple of fractals from presets:

```julia
γ = CantorSet()
Δ = Sierpinski();
```

### Gauss quadrature
Gauss quadrature is only possible on fractals where $\Gamma\subset\mathbb{R}$. So we can do this on the Cantor set $\gamma$, an $N$-point rule can be constructed like so:

```julia
N = 51
x,w = gauss_quad(γ,N);
scatter(x,ones(length(x)),markersize=1)
```

Notice how not all of the quadrture points lie inside of the fractal $\gamma$, there is clearly some influence from $\gamma$. So, to approximate
$$
\int_\gamma \sin(x)\mathcal{H}^d(x)
$$
we can write

```julia
w'*sin.(x)
```

## Barycentre rule
To use the barycentre rule, with meshwidth parameter $h>0$, type:

```julia
h = 0.01
x,w = barycentre_rule(γ,h)
scatter(x,ones(length(x)),markersize=1)
```

```julia
w'*sin.(x)
```

The barycentre rule will converge slower than Gauss, but has the advantage that it can be applied to fractals in any dimension. For example:

```julia
x,w = barycentre_rule(Δ,h)
```

```julia
scatter([x_[1] for x_∈x],[x_[2] for x_∈x],markersize=1)
```

To approximate
$$
\int_\Delta \sin(|x|) \mathcal{H}^d(x)
$$
we can type

```julia
using LinearAlgebra
w'*sin.(norm.(x))
```

```julia
N = 1000
x,w = chaos_quad(Δ,N);
```
